```python
import requests
import feedparser

url = "http://www.moneyworks4me.com/company/news/latest-stock-news-rss/corporate-news"  # example RSS feed URL
response = requests.get(url, verify=False)
feed = feedparser.parse(response.content)
print(feed)
```

```python
# Example: Reading S&P 500 RSS feed using the scraper
from scrapers.rss_scraper import RSSFetcher, Scraper
from datetime import date

# Initialize fetcher and scraper
fetcher = RSSFetcher(config_path="config/config.json")
fetcher.fetch_all()

# Get entries from S&P 500 feed
sp500_entries = fetcher.get_entries("SP500_DAILY_INSIGHTS")
print(f"Found {len(sp500_entries)} entries from S&P 500 feed\n")

# Display first few entries
for idx, entry in enumerate(sp500_entries[:3], 1):
    print(f"{idx}. {entry.get('title', 'No title')}")
    print(f"   Published: {entry.get('published', 'N/A')}")
    print(f"   Link: {entry.get('link', 'N/A')}")
    print(f"   Description: {entry.get('description', 'N/A')[:100]}...")
    print()
```

```python
# Example: Collect S&P 500 feed with full markdown content
scraper = Scraper(fetcher)

# Collect only S&P 500 feed
sp500_collection = scraper.collect_feed("SP500_DAILY_INSIGHTS")

print(f"Collected {len(sp500_collection['articles'])} articles from S&P 500")
print("\nArticles with markdown content:")
for article in sp500_collection['articles']:
    print(f"\n- {article['title']}")
    print(f"  Published: {article['published']}")
    print(f"  Link: {article['link']}")
    print(f"  Content preview: {article['content'][:150]}...")
```
